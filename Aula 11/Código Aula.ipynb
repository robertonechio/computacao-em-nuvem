```
from googletrans import Translator

# Initialize the translator
translator = Translator()

# Text to translate
text = "Hello, world! How are you today?"

# List of target languages (e.g., Portuguese, Spanish, French, German, Italian)
target_languages = ["pt", "es", "fr", "de", "it"]
source_language = "en"

print("Texto original:", text)

# Perform the translation for each language
for lang_code in target_languages:
    translation = translator.translate(text, src=source_language, dest=lang_code)
    print(f"Tradução ({lang_code}):", translation.text)
```

```
from PIL import Image
import requests
from io import BytesIO
from transformers import pipeline

# Carregar modelo pré-treinado
classificador = pipeline("image-classification", model="microsoft/resnet-50")

# Imagem de exemplo
# url_imagem = "https://upload.wikimedia.org/wikipedia/commons/9/9a/Pug_600.jpg"
# url_imagem = "https://www.tensorflow.org/images/colab_logo_32px.png" # Trying a different URL
url_imagem = "https://www.pngall.com/wp-content/uploads/2016/03/Dog-PNG-Image.png" # PNG image of a dog

# Fetch the image content, bypassing SSL verification
response = requests.get(url_imagem, verify=False)

# Check if the request was successful
if response.status_code == 200:
    try:
        # Open the image from the bytes content
        imagem = Image.open(BytesIO(response.content))
        # imagem.show() # This might not work directly in Colab without extra setup

        # Classificar a image
        # Pass the PIL image object as a list to the pipeline
        resultados = classificador([imagem])


        print("Resultados da classificação:\n")
        for r in resultados:
            print(f"{r['label']} - Confiança: {r['score']:.2%}")
    except UnidentifiedImageError:
        print(f"Error: Could not identify image from URL: {url_imagem}")
    except Exception as e:
        print(f"An error occurred: {e}")
else:
    print(f"Error: Failed to fetch image from {url_imagem}. Status code: {response.status_code}")
```
